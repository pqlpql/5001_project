```python
path_train = '/content/drive/MyDrive/Colab Notebooks/pql/kaggle_kidney_data/train.csv'
df=pd.read_csv(path_train)
```

```python
# fill null data
df[df.isnull().values==True]
```

```python
df = df.fillna(df.mean())
```

```python
# drop unuseful column
df.drop(['id'], axis=1, inplace=True)
```

```python
# split label
y = df.loc[:,'label']
df = df.drop(['label'],axis=1)
```

```python
# scaling
from sklearn import preprocessing
# Get column names first
names = df.columns
# Create the Scaler object
scaler = preprocessing.StandardScaler()
# Fit your data on the scaler object
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=names)
```

```python
scaled_df
```

```python
scaled_df.describe()
```

```python
# from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
"""
  over sampling
  but it is unuseful, so finally this data is not uesd
"""
# extract label
smt = SMOTE(sampling_strategy={1: 100,0:100})
# smo = SMOTE(sampling_strategy=0.4,random_state=42)
X_smo, y_smo = smt.fit_resample(scaled_df, y)
```

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import StratifiedKFold 
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
# see original score
model = RandomForestClassifier()
x_train, x_test, y_train, y_test = train_test_split(X_smo, y_smo, test_size = 0.1, random_state = 0)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(accuracy_score(y_test , y_pred)*100)
```

```python
"""
  adjust the parameter using gridsearchcv
  including
    1.max_depth
    2.n_estimators
    3.max_features
"""
param_test = {'max_features':range(2,12,2),'max_depth':range(2,12,2),'n_estimators':range(50,500,100)}
kflod = StratifiedKFold(n_splits=5, shuffle = True,random_state=7)
gsearch2 = GridSearchCV(estimator = RandomForestClassifier(oob_score=True),param_grid = param_test, scoring='accuracy',n_jobs = -1, cv=kflod)
gsearch2.fit(X_smo,y_smo)
print("Best: %f using %s" % (gsearch2.best_score_,gsearch2.best_params_))
```

```python
"""
  data preprocessing in test data
"""
def data_preprocessing(df):
  df.drop(['id'], axis=1, inplace=True)
  df = df.fillna(df.mean())
  # Get column names first
  names = df.columns
  # Create the Scaler object
  scaler = preprocessing.StandardScaler()
  # Fit the data on the scaler object
  scaled_df = scaler.fit_transform(df)
  scaled_df = pd.DataFrame(scaled_df, columns=names)
  return scaled_df
```

```python
"""
  model in test, output submission.csv
"""
"""
  after gridsearchcv,the accuracy did not rise obviously, to prevent overfitting, the paras are not set too many
"""
test_path = '/content/drive/MyDrive/Colab Notebooks/pql/kaggle_kidney_data/test.csv'
test_x = pd.read_csv(test_path)
test_x = data_preprocessing(test_x)
rfc = RandomForestClassifier(n_estimators=200,oob_score=True)
rfc.fit(scaled_df, y)
test_y_pred = rfc.predict(test_x)
output=pd.DataFrame(test_y_pred.astype(np.int32),columns=['label'])
output['id']=range(output.shape[0])
output=output.loc[:,['id','label']]
output.to_csv('/content/drive/MyDrive/Colab Notebooks/pql/kaggle_kidney_data/submission.csv',index=False)
```
